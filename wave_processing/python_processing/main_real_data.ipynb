```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
GRAVITY = 9.81
NUM_COLUMNS = 8000
PERCENT_SHOWN = .025 #Percentage of data shown in graphs
```

```python
df = pd.read_csv('OrginalData.csv')
df.shape
print(df.head)

SAMPLING_FREQUENCY = 16 #16Hz -
frequency_resolution = SAMPLING_FREQUENCY / NUM_COLUMNS #frequncy interval for each frequency bin
freq_domain = np.linspace(0,(NUM_COLUMNS-1)*frequency_resolution, NUM_COLUMNS) #x-axis for frequency spectrum
```

```python
#convert to vertical accelerations
vertical_earth_accelerations = []

for i in range(0,NUM_COLUMNS):
    vertical_earth_accelerations.append(GRAVITY *(1 
                                                  - math.sqrt(df[' Az (m/s^2)'][i]**2 
                                                  + df[' Ay (m/s^2)'][i]**2
                                                 + df[' Ax (m/s^2)'][i]**2))) 
```

```python
#linearly detrend the vertical accelerations
sumX, sumY, sumXY, sumXX = 0,0,0,0

for i in range(0,NUM_COLUMNS):
    sumX += i+1;
    sumY += vertical_earth_accelerations[i]  
    sumXY += (i+1) * vertical_earth_accelerations[i]
    sumXX += (i+1) * (i+1)
    
slope = (NUM_COLUMNS * sumXY - sumX * sumY) / (NUM_COLUMNS * sumXX - sumX * sumX)
intercept = (sumY - slope * sumX) / NUM_COLUMNS

for i in range(0,NUM_COLUMNS):
    vertical_earth_accelerations[i] -= (slope * (i+1) + intercept) 
    
```

```python
mean = np.mean(vertical_earth_accelerations)
std_dev = np.std(vertical_earth_accelerations)
print(mean)
print(std_dev)
```

```python
plt.figure(figsize=(10,5))   
plt.plot(vertical_earth_accelerations)

# Set the plot title and axis labels
plt.title('detrended earth vertical acceleration')
plt.xlabel('Index')
plt.ylabel('vertical_acceleration')

plt.axhline(y=mean, color='blue', linestyle='--')
plt.axhline(y=(mean+3*std_dev), color='black', linestyle='--')
plt.axhline(y=(mean-3*std_dev), color='black', linestyle='--')

# Show the plot
plt.show()
```

```python
from scipy.interpolate import CubicSpline
# Convert the list to a NumPy array for easier manipulation
y = np.array(vertical_earth_accelerations)

# Identify outlier indices
outlier_indices = np.where(np.abs(y - mean) > 3 * std_dev)[0]    

# Create valid indices
valid_indices = np.arange(len(vertical_earth_accelerations))
valid_indices = np.delete(valid_indices, outlier_indices)
        
#Sort the valid indices
valid_indices_sorted = np.sort(valid_indices)

# Perform cubic spline interpolation on valid data
cs = CubicSpline(valid_indices_sorted, y[valid_indices_sorted])

# Interpolate missing values                     
interpolated_values = cs(outlier_indices)

print(interpolated_values)

# Replace interpolated values in the original list
for i, index in enumerate(outlier_indices):
    y[index] = interpolated_values[i]
```

```python
#Apply kaiser bessel window
from scipy import signal
from scipy.fft import fft, fftshift

window = signal.windows.kaiser(512, beta = 0.5, sym = True)
appliedSegments = list(map(lambda segment: segment * window, segments))
```

```python
# FFT Transform and average the values
fftSegments = list(map(lambda x: np.abs(fft(x)) / len(x), appliedSegments))

# Take only half
fftSegments = list(map(lambda x: x[:len(x)//2], fftSegments))
#average the segments
fftSegmentAvg = [sum(column)/17 for column in zip(*fftSegments)]
print(len(fftSegmentAvg))
plt.plot(fftSegmentAvg)
plt.xlabel('frequency (Hz)')  # Replace 'X-axis Label' with your desired label for the x-axis
plt.ylabel('m^2/((Hz)s^4)')
plt.show()
```

```python
#convert to the displacement spectra by dividing by the frequency to the fourth power.

displacement_spectra = list(map(lambda x, y: x/(y**4) if y !=0 else 0, fftSegmentAvg,freq_domain))

#y/x^4   the frequency that correrpsonds to the y value

plt.plot(displacement_spectra)
plt.xlabel('frequency (Hz)')  # Replace 'X-axis Label' with your desired label for the x-axis
plt.ylabel('m^2/Hz')
plt.show()
```
